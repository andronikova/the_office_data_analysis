# Linear regression

```python
import statsmodels.formula.api as smf
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.30)
```

```python
def RMSE(y, y_pred):
    SME = sum((y - y_pred)**2)
    
    return round(math.sqrt(SME/len(y)),2)
```

```python
formula = 'rating ~ '

for value in rating_high_corr.index:
    if value != 'rating':
        formula += value + " + "

formula = formula[:-2]
formula
```

```python
formula = 'rating ~ Michael_words + double + unhappy + director_Paul_Feig + writer_Greg_Daniels + sigh + Jan_words + what_she_said + writer_Paul_Lieberstein + Pam_about_Michael + writer_Michael_Schur + director_Greg_Daniels + director_Harold_Ramis + Jim_about_Pam + Holly_words + director_Ken_Kwapis + writer_Jennifer_Celotta + Jim_about_Dwight + director_Steve_Carell + director_Ed_Helms + Kevin_words + halloween + director_Matt_Sohn + writer_Steve_Hely + director_John_Krasinski + Gabe_words + Darryl_words + Andy_words + Erin_words + Nellie_words '
```

```python
model = smf.ols(formula, data=train)
results = model.fit()
results.summary()
```

```python
prediction = results.predict(test.drop(columns='rating', axis=1))

sns.scatterplot(x=test.rating, y=prediction)
```

```python
RMSE(test.rating.values, prediction.values)
```
